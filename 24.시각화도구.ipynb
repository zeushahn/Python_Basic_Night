> 년도 Data들이 object로 되어 있다.   
0 번 index를 삭제

```python
# index 0을 삭제
df.drop(
    0,
    axis='rows',
    inplace=True
)

df.head()
```

```python
# Index 재정리
df.reset_index(
    inplace=True,
    drop=True # index column 생성 방지 
)
df.head()
```

```python
# NaN 값 정리
# NaN을 전(Prev) 데이터로 채움 

df = df.ffill()
df.head()
```

### 년도별 컬럼에 대해 Object를 숫자로 변환

```python
# '-' 을 NaN으로 변경
df.replace('-', np.nan, inplace=True)
df.head()
```

```python
# 년도별 컬럼의 값을 숫자로 변환하고 숫자로 변환되지 않는 값은 NaN으로 처리 
df.iloc[:,2:] = df.iloc[:,2:].apply(pd.to_numeric, errors='coerce')

# NaN을 0으로 전환 
df.fillna(0, inplace=True)
```
